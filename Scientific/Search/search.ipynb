جستجو در گراف (Graph Search)
        </font>
        <p></p>
    الگوریتم های جست و جو در گراف پایه‌ای‌ترین الگوریتم ها در هوش مصنوعی هستند. در همهٔ آن‌ها برای مسئله‌های هوش مصنوعی یک گراف می‌سازیم که هر راس آن نمایشگر یک وضعیت (state) از بازی مورد نظر ماست و از یک راس به یک راس دیگر یال می گذاریم اگر بتوان با انجام یک عملیات از این وضعیت به آن وضعیت رسید.<br>
    حال ما در این گراف یک وضعیت اولیه داریم و هدفمان این است که با انجام تعدادی عملیات به یک وضعیت برد برسیم. و اینجاست که الگوریتم های جست‌و‌جو وارد کار می شوند و به ما در پیدا کردن این دنباله از عملیات ها کمک می کنند.<br>
        می‌توان از الگوریتم های Search به DFS , BFS, Bidirectional Search, Depth Limited Search اشاره کرد که ما در این جا دو تا از آن ها را که نسبت به بقیه مهم تر هستند، توضیح می دهیم.<br>
        هم‌چنین یکی دیگر از الگوریتم های پرکاربرد الگوریتم Uniform Cost Search است که کاربرد آن در گراف هایی است که هزینهٔ طی کردن یال‌ها در آن‌ها متفاوت است و توصیه می کنیم این الگوریتم را هم مطالعه کنید!

جستجوی اول عمق (DFS)
        </font>
        <p></p>
        می‌توان گفت یکی از پایه‌ای ترین الگوریتم های جستجو در گراف، الگوریتم جست‌و‌جوی اول عمق (DFS) است. نحوهٔ کار این الگوریتم به شکل زیر است :<br>
        در هر لحظه از زمان ما یک مسیر از ریشه به یک راس مانند $v$ داریم و یک مجموعهٔ $s$ از راس هایی که قبلا دیده شده اند. حال در هر مرحله اگر یک همسایهٔ $v$ مانند $u$ وجود داشت که در $s$ نبود ما $u$ را به مسیر اضافه می کنیم و آن را درون مجموعهٔ $s$ قرار می  دهیم. در غیر این صورت تمام همسایه های $v$ در $s$ آمده‌اند؛ پس با ادامه دادن این مسیر نمی‌توانیم به یک راس هدف برسیم و $v$ را از انتهای مسیر حذف می‌کنیم و الگوریتم را ادامه می‌دهیم. در ادامه پیاده سازی این الگوریتم را روی مسئلهٔ پازل ۸ تایی می بینید:   
    </font>
</div>

```python
%%writefile codes/search.py
import driver
from state import *

def search(start_state):

    explored, stack = set(), list([State(start_state, None, None, 0, 0, 0)])
    while stack:
        #reading from stack
        node = stack.pop()

        explored.add(node.map)

        if node.state == driver.goal_state:
            driver.goal_node = node
            return stack

        neighbors = reversed(driver.expand(node))

        for neighbor in neighbors:
            if neighbor.map not in explored:
                #adding to dfs stack
                stack.append(neighbor)
                explored.add(neighbor.map)

                if neighbor.depth > driver.max_search_depth:
                    driver.max_search_depth += 1

        if len(stack) > driver.max_frontier_size:
            driver.max_frontier_size = len(stack)
```

<div id="bfs" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
جستجوی اول سطح (BFS)
        </font>
        <p></p>
        یکی دیگر از الگوریتم های پایه‌ای جست‌و‌جو در گراف، جست‌و‌جوی اول سطح یا BFS است. ایدهٔ کلی این الگوریتم این است که بر خلاف DFS سعی می‌کند به ازای هر راس کوتاه ترین مسیر از راس شروع به آن را پیدا کند تا هزینهٔ کم‌تری را برای رسیدن به وضعیت نهایی صرف کنیم. برای پیاده سازی این ایده در هر مرحله از زمان یک صف داریم که به ترتیب شامل راس هایی است که قرار است الگوریتم ببیند. حال در هر مرحله ما راسی که در سر صف قرار دارد را از سر صف حذف می کنیم و تمام همسایه هایش که دیده نشده‌اند را به صف اضافه می کنیم. بدین ترتیب ما به ترتیب فاصله از راس شروع راس ها را می بینیم و می توانیم کوتاه ترین مسیر به وضعیت نهایی را پیدا کنیم. در ادامه می‌توانید پیاده سازی این الگوریتم را ببینید: 
    </font>
</div>

```python
%%writefile codes/search.py
from collections import deque
import driver
from state import *

def search(start_state):

    explored, queue = set(), deque([State(start_state, None, None, 0, 0, 0)])

    while queue:
        #reading from queue
        node = queue.popleft()

        explored.add(node.map)

        if node.state == driver.goal_state:
            driver.goal_node = node
            return queue

        neighbors = driver.expand(node)

        for neighbor in neighbors:
            if neighbor.map not in explored:
                #adding to bfs queue
                queue.append(neighbor)
                explored.add(neighbor.map)

                if neighbor.depth > driver.max_search_depth:
                    driver.max_search_depth += 1

        if len(queue) > driver.max_frontier_size:
            driver.max_frontier_size = len(queue)
```

  جستجوی مطلع (Informed Search)
        </font>
        <p></p>
        فرض کنید شما در تهران هستید و می خواهید با طی کردن یک مسیر کوتاه به شیراز بروید و نقشهٔ جاده های ایران هم رو به رویتان است. در این صورت احتمالا شما الگوریتم BFS را اجرا نمی کنید و به دنبال مسیری می گردید که تقریبا  مساوی خط راست بین این دو نقطه باشد. یعنی به صورت دقیق‌تر به ازای هر شهر یک فاصلهٔ تخمینی تا شیراز به دست می‌آورید که همان فاصلهٔ اقلیدسی بین شیراز و آن شهر است و در هر مرحله به شهری می‌روید که فاصلهٔ تخمینی کم‌تری دارد. البته ممکن است مسیری که به دست می‌آید کوتاه‌ترین مسیر نباشد اما در این مثال و خیلی مثال‌های دیگر می‌توانیم با همین روش کوتاه‌ترین مسیر را با صرف هزینهٔ بسیار کم‌تری به دست آوریم. <br>
        حال سعی می‌کنیم دسته‌ای از الگوریتم‌ها را که پایهٔ آن‌ها پیدا کردن مسیر به کمک فاصلهٔ تخمینی تا وضعیت نهایی است به شما معرفی کنیم. <br> 
         شیوهٔ کلی کار این الگوریتم‌ها بدین گونه است که با توجه به اطلاعاتی که از مسئله داریم، برای هر وضعیت از بازی مانند $v$ یک مقدار $h(v)$ تعریف می‌کنیم که فاصلهٔ حدودی تا رسیدن به یکی از وضعیت های هدف است.<br>
        حال ما هر چه بتوانیم این تابع $h$  را بهتر تخمین بزنیم می توانیم مسئله را هم بهتر حل کنیم.
        برای مثال فرض کنید یک کشور داریم که از تعدادی شهر و جاده تشکیل شده است و ما می خواهیم کم ترین فاصله از یک شهر دلخواه را با یکی از شهر هایی که دارای پمب بنزین است به دست آوریم. ما می توانیم تابع $h(v)$ را بدین شکل تعریف کنیم که برابر کم ترین فاصله اقلیدسی شهر $v$ و یکی از شهر های دارای پمپ بنزین باشد.<br>
        حال یک الگوریتم خوب با توجه به این تابع این است که ما در هر لحظه از زمان وارد راسی شویم که دارای کم ترین مقدار فاصلهٔ تقریبی است و این کار را تا جایی ادامه دهیم که به یک شهر دارای پمپ بنزین برسیم.
        این الگوریتم یک مسیر نسبتا کوتاه به یک شهر دارای پمپ بنزین به دست می‌آورد. البته توجه کنید که این مسیر لزوما کوتاه‌ترین مسیر نیست.
    </font>

<div id="A*" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
الگوریتم $A^*$
        </font>
        <p></p>
        فرض کنید که $h^*(v)$ فاصلهٔ حقیقی راس v از هدف باشد
        و h که در بخش بالا معرفی کردیم دو ویژگی زیر را داشته باشد :<br>
        ۱ - به ازای هر v، $h(v) \le h^*(v)$ برقرار باشد. <br> 
        ۲ - به ازای هر راس v و هر یالی که از v خارج می‌شود و طول c دارد و به راس u می‌رود، داشته باشیم: $h^*(v) \le c + h^*(u)$ <br>
        حال الگوریتم $A^*$ به شکل زیر کار می کند:  <br>
        در هر مرحله از زمان ما تابع $g$ را به ازای هر راس به شکل زیر تعریف می کنیم : $g(v) = h(v) + f(v)$ که در آن $f(v)$ هزینهٔ کوتاه ترین مسیری است که تا به حال به راس $v$ پیدا کرده‌ایم. حال در هر مرحله از زمان راسی که تابع $g$ در آن کم‌ترین مقدار را دارد در نظر می‌گیریم و فاصلهٔ اصلی این راس را برابر $g$ قرار می‌دهیم و تابع $f$ را برای همسایه های $v$ دوباره حساب می‌کنیم و این روند را تا زمانی که تمام فاصله‌ها حساب شوند، ادامه می‌دهیم.<br>
        می‌توان اثبات کرد که این الگوریتم همیشه به ازای هر راس کوتاه‌ترین فاصله را به دست می‌آورد که اثبات آن را به خواننده واگذار می‌کنیم. البته اگر این مسئله برای شما کمی پیچیده است نگران نباشید؛ چون این الگوریتم کمی پیچیده است. اما همیشه سعی کنید تابع تخمینی که ارائه می‌دهید دو ویژگی بالا را داشته باشد. اگر کمی به آن ویژگی‌ها دقت کنید خیلی منطقی به نظر می رسند!<br>
        حال به بررسی کاربرد‌های این الگوریتم  می‌پردازیم و برای تست کردن این الگوریتم آن را روی مسئلهٔ پازل ۸ تایی پیاده سازی می‌کنیم و آن را با الگوریتم های قبلی مقایسه می‌کنیم. شما در اینجا می‌توانید تابع $h$ را خودتان تغییر دهید و میزان تغییر محاسبات را مشاهده کنید.
    </font>
</div>

```python
%%writefile codes/search.py
import driver
from state import *
from heapq import heappush, heappop, heapify
import itertools

#you can heuristic function here:
def h(state):
    return sum(abs(b % driver.board_side - g % driver.board_side) + abs(b//driver.board_side - g//driver.board_side)
               for b, g in ((state.index(i), driver.goal_state.index(i)) for i in range(1, driver.board_len)))

def search(start_state):

    global max_frontier_size, goal_node, max_search_depth

    explored, heap, heap_entry, counter = set(), list(), {}, itertools.count()

    key = h(start_state)

    root = State(start_state, None, None, 0, 0, key)

    entry = (key, 0, root)

    heappush(heap, entry)

    heap_entry[root.map] = entry

    while heap:
        #finding best node
        node = heappop(heap)

        explored.add(node[2].map)

        if node[2].state == driver.goal_state:
            driver.goal_node = node[2]
            return heap

        neighbors = driver.expand(node[2])

        #updating neighbors values
        for neighbor in neighbors:

            #calculating values
            neighbor.key = neighbor.cost + h(neighbor.state)

            entry = (neighbor.key, neighbor.move, neighbor)

            if neighbor.map not in explored:

                heappush(heap, entry)

                explored.add(neighbor.map)

                heap_entry[neighbor.map] = entry

                if neighbor.depth > driver.max_search_depth:
                    driver.max_search_depth += 1

            elif neighbor.map in heap_entry and neighbor.key < heap_entry[neighbor.map][2].key:

                hindex = heap.index((heap_entry[neighbor.map][2].key,
                                     heap_entry[neighbor.map][2].move,
                                     heap_entry[neighbor.map][2]))

                heap[int(hindex)] = entry

                heap_entry[neighbor.map] = entry

                heapify(heap)

        if len(heap) > driver.max_frontier_size:
            driver.max_frontier_size = len(heap)
```

        احتمالا با این مسئله آشنایی داشته باشید؛ هدف ما در این مسئله، قرار دادن $n$ وزیر در یک صفحه‌ی $n\times n$ شطرنج است، به طوری که یک‌دیگر را تهدید نکنند. به عنوان مثال، شکل زیر یک راه حل برای این مسئله با مقدار $n=8$ است:

        حال فرض کنید که می‌خواهیم این مسئله را مدل‌سازی کرده و با کمک کامپیوتر حل کنیم. اگر دقت کرده باشید، بر خلاف مسئله‌های قبلی (مثلا پیدا کردن نزدیک‌ترین مسیر از یک شهر به شهر دیگر) در این مسئله <b> مسیر رسیدن به هدف برای ما اهمیتی ندارد</b>. به بیان دیگر، ما تنها به دنبال چینشی از مهره‌های وزیر هستیم که یک‌دیگر را تهدید نکنند و مسیر پیموده شده برای رسیدن به این حالت اولویت ما نیست. <br>
        برای حل این گونه مسائل که در آن‌ها تنها به دنبال رسیدن به یک حالت قابل قبول هستیم، از الگوریتم‌های جستجوی محلی استفاده می‌شود. از این الگوریتم‌ها می‌توان به Local Beam Search, Simulated Annealing و Genetic Algorithms اشاره کرد. در ادامه الگوریتم‌های ژنتیک را توضیح می‌دهیم.

الگوریتم‌های ژنتیک (Genetic Algorithms)
        </font>
		<p></p>
		<hr>
الگوریتم‌های ژنتیک الهام‌گرفته‌شده از تکامل و انتخاب طبیعی در طبیعت بوده و به خصوص برای مسائل بهینه‌سازی و جستجو با فضای حالت بزرگ کاربردی هستند.<br>
        همان طور که گفته شد، در مسائل جستجوی محلی هدف اصلی ما پیدا کردن یک جواب برای مسئله و نه لزوما راه رسیدن به آن است. در الگوریتم‌های ژنتیک، برای مدل‌سازی هر حالت از مسئله از مفهوم ژن استفاده می‌کنیم. هر ژن را می‌توان یک رشته از اعداد در نظر گرفت که یک حالت از مسئله را مدل‌سازی کرده است. به عنوان مثال، در مسئلهٔ n-وزیر، می‌توانیم به ازای هر یک از ستون‌ها، شمارهٔ سطری که وزیر در آن قرار دارد را در نظر گرفته و با یک رشتهٔ n تایی، یک حالت از جواب دادن به مسئله را مدل کنیم. دقت کنید که این ژن‌ها تنها حالت‌های ممکن مسئله بوده و لزوما یک جواب قابل قبول نبوده و پیدا کردن چنین ژنی هدف الگوریتم است. در شکل زیر ژن‌های مربوط به دو چینش وزیرها را مشاهده می‌کنید. (شمارش ردیف‌ها و ستون‌ها از سمت چپ و بالا شروع می‌شود)<br>
        <img src = "images/chess_gen.png" style="width:80%">
        در الگوریتم‌های ژنتیک، یک جمعیت (population) از جواب‌های ممکن داریم که یک راه حل احتمالی را نشان می‌دهند. در هر دور (iteration/generation) با استفاده از تکنیک‌هایی، این جمعیت به روزرسانی می‌شود.<br>
        به عنوان مثال، در مسئلهٔ n-وزیر، می‌توان هر جواب احتمالی را یک چینش از n مهرهٔ وزیر در صفحه‌ی شطرنج در نظر گرفت. بنابراین، یک جمعیت به مجموعه‌ای از چند چینش مختلف مهره‌ها در صفحه تبدیل شده که در هر دور از مسئله، تلاش داریم آن را به روز رسانی کنیم.<br>
        می‌توان مراحل این الگوریتم‌ها را به چند گام زیر خلاصه کرد:
        <ol>
        <li>ایجاد یک جمعیت اولیه به طور تصادفی
        <li>محاسبهٔ امتیاز (fitness) هر یک از اعضای جمعیت
        <li>انتخاب تعدادی از جواب‌ها برای ترکیب و تولید نسل جدید
        <li>ترکیب جواب‌های انتخاب شده و افزودن نتیجه به جمعیت
        <li>تکرار از گام ۲، تا زمانی که به یک جواب مناسب رسیده یا الگوریتم به تعداد دفعات مشخصی تکرار شده باشد. در مسئلهٔ فعلی، تکرار الگوریتم را تا جایی ادامه می‌دهیم که یک جواب قابل قبول برای مسئله در بین جمعیت وجود داشته باشد.</ol>
        در ابتدا لازم است که با چند اصطلاح کلیدی در این الگوریتم‌ها آشنا شویم:<ol>
        <li>ترکیب (crossover): همان طور که گفتیم در هر گام از الگوریتم، دو تا از جواب‌های موجود در جمعیت را با یک‌دیگر ترکیب کرده و جواب جدید ایجاد می‌کنیم. با این کار فرزندی تولید شده که ترکیبی از ویژگی‌های والدینش را داشته و این کار به روش‌های مختلف و با توجه به مسئله می‌تواند انجام شود. مثلا در مسئهٔ فعلی، می‌توان برای ترکیب دو ژن والدین، نیمهٔ سمت چپ و راست آن‌ها را جدا کرده و به یک‌دیگر متصل کرد. این شیوهٔ ترکیب کردن شبیه این است که دو صفحهٔ شطرنج که در هر ستونش دقیقا یک وزیر دارد را به طور عمودی از وسط بریده، نیمه‌های چپ و راست این دو صفحه را به یکدیگر متصل کنیم.
        <li>جهش (mutation): پس از ترکیب والدین و تولید جواب جدید، به احتمال کمی این جواب یک جهش خواهد داشت. به این معنی که یکی از ژن‌هایش تغییر کرده و به جواب جدیدی تبدیل می‌شود. این کار را برای کم کردن احتمال گیر کردن در کمینهٔ محلی انجام می‌دهیم، چون با ایجاد تغییری کوچک ولی تصادفی، احتمال بررسی کردن نقاط مختلف فضای حالت بالاتر می‌رود.<br>
        در شکل زیر، تاثیر اجرای ترکیب و جهش بر روی دو ژن این مسئله را مشاهده می‌کنید:
        <img src = "images/chess_cross_mutation.png" style="width:50%">
        <li>انتخاب (selection): یکی از نکات مهم در الگوریتم‌های ژنتیک، این است که باید در هر دور تعدادی از ژن‌ها را به عنوان والدین انتخاب کنیم. برای این کار تلاش می‌کنیم تا میزان مناسب بودن هر ژن را، با توجه به مسئله، تخمین بزنیم که این کار با تابعی به نام fitness function انجام می‌شود. در ابتدا میزان شایستگی هر یک از جواب‌ها را با استفاده از تابعی به نام fitness function سنجیده و سپس تعدادی از جواب‌ها را به صورت تصادفی انتخاب می‌کنیم. با توجه به این که هرچه امتیاز شایستگی یک جواب بیشتر باشد جواب بهتری فرض می‌شود، احتمال انتخاب شدن هر یک جواب‌ها با افزایش شایستگی‌اش بیشتر می‌شود. یک راه معمول برای مشخص کردن احتمال انتخاب شدن یک جواب به صورت زیر است:
        $$chance(i) = \frac{fitness(i)}{\sum_{k \in P}fitness(k)}$$
        همان‌طور که مشاهده می‌شود، در این فرمول احتمال انتخاب شدن هر یک از ژن‌ها برای تولید نسل بعدی متناسب با مقدار fitness function آن ژن است. با دخیل کردن شانس در هنگام انتخاب این ژن‌ها، احتمال گیر کردن در local minima را نیز کمتر می‌کنیم. (چون گاهی اوقات ممکن است ژن‌هایی که چندان مناسب نیستند را هم انتخاب کرده و بدین ترتیب قسمت‌های مختلفی از فضای حالت را بررسی کنیم)</ol>
        انتخاب تابع fitness:<br>
        همان طور که گفتیم، با استفاده از تابع fitness باید بتوانیم به ازای هر حالت (ژن) تخمینی از خوب بودن آن را به دست آوریم. این موضوع به تعریف مسئله بستگی داشته و با توجه به این موضوع، می‌توان توابع گوناگونی را انتخاب کرد. به عنوان مثال، در مسئله‌ی فعلی می‌توانیم در هر حالت، تعداد جفت‌وزیرهایی که یک‌دیگر را تهدید می‌کنند را از تعداد کل جفت وزیرهای ممکن کم کرده و به عنوان تابع fitness قرار دهیم. به این حالت توجه کنید:
    <img src = "images/chess_fitness.png" style="width:30%">
    با توجه به این که کلا ۸ وزیر داریم، تعداد کل جفت‌های ممکن برابر است با: $C(8, 2) = 28$. همچنین، مشاهده می‌شود که ۴ جفت وزیر یک‌دیگر را تهدید می‌کنند. یعنی مقدار تابع fitness برای این حالت برابر با $28-4=24$ است. <br>
    همان طور که می‌بینید، به ازای چینش‌هایی که جفت وزیرهای کمتری یکدیگر را تهدید می‌کنند مقدار این تابع بیشتر شده و احتمال انتخاب شدن ژن مربوطه نیز بالا می‌رود. با چند بار اجرای کد زیر می‌توانید چند نمونهٔ مختلف از روند این الگوریتم را مشاهده کنید(اگر با پیام خطا مواجه شدید از منوی Kernel گزینهٔ Restart را انتخاب کنید و دوباره اجرا کنید).
